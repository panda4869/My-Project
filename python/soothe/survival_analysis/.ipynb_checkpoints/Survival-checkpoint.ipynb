```python
from lifelines import KaplanMeierFitter 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
data=pd.read_csv('data.csv')
```

```python
data.head()
```

```python
data['observe']=np.where(data.not_book_interval >=90,0,1)
```

```python
#data.loc[(data.not_book_interval >90),'not_book_interval']=180
```

```python
data.head()
```

```python
dataset=data.drop(['user_id','interval_between_bookings'],axis=1)
```

```python
dataset.head()
```

```python
index=random.sample(range(dataset.shape[0]),10000)
kmf=KaplanMeierFitter()
kmf.fit(durations=dataset.not_book_interval[index],event_observed=data.observe[index],label='Probability Curve')
#plt.figure(figsize=(12,10))
ax=plt.subplot(111)
plt.rcParams["figure.figsize"] = (12,10)
p=kmf.plot(ci_force_lines=False,title='Probability to Book Again within 90 Days',ax=ax)
plt.ylim([0.0,1.0])
vals=ax.get_yticks()
ax.set_yticklabels(['{:3.0f}%'.format(x*100) for x in vals])
plt.xlabel('Booking Intervals (Days)')
plt.xlim([0,90])


plt.show()
```

```python
kmf.survival_function_
```

```python
ax=plt.subplot(111)
for r in data['rank'].unique():
    ix=data['rank']==r
    kmf=KaplanMeierFitter()
    kmf.fit(data['not_book_interval'].ix[ix],data['observe'].ix[ix],label=r)
    ax=kmf.plot(ax=ax)

plt.xlim([0,90])
plt.ylim([0.0,1.0])
vals=ax.get_yticks()
ax.set_yticklabels(['{:3.0f}%'.format(x*100) for x in vals])
plt.show()
```
